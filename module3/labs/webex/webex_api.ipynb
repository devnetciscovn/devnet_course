```python
import requests
from credentials import *

def send_message(email, text):
    url = base_url  + "/v1/messages"

    headers = {
        "Authorization": f"Bearer {token}"
    }

    body = {
        "toPersonEmail": email,
        "text": text
    }
    
    response = requests.post(url, headers = headers, json = body)
    print(email, response.status_code)
    
```
